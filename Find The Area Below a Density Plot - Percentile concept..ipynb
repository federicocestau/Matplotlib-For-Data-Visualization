# kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram.

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# Set style and font scale using Seaborn 
sns.set_theme(style='whitegrid', font_scale = 1.75)
# set figure size
plt.figure(figsize=(10, 10)) 

# Draw Density Curve
# fill=True shades the area under the curve  
sns.kdeplot(heights['mean'], linewidth=2, fill=True)

# Vertical line at the peak value - for reference 
plt.axvline(x=64.57, linewidth=3, color='red', linestyle='--')

# title & labels
plt.title("FIG 1. Density Curve for High School Heights", pad=20)
plt.xlabel("Height (inches)", labelpad=20)
plt.ylabel("Density", labelpad=20)
```

Area Under Density Curve We know that the sum of probabilities for a variable is 1. And as mentioned earlier, the density curve shows the probabilities for all the values of a variable.

Therefore, the total area under the density curve always equals 1.

This simple fact is immensely useful and has many practical applications.

Imagine if someone asks you - what percentage of data values fall within a specific interval. You can answer such queries by measuring the partial area under the curve.

```python
plt.figure(figsize=(16, 10))

# Draw the density curve with it's area shaded
sns.kdeplot(heights['mean'], linewidth=2, fill=True)
# Invoke kdeplot() again to get reference to axes 
ax = sns.kdeplot(heights['mean'])

# Below code to shade partial region is from 
# https://stackoverflow.com/a/49100655

# Get all the lines used to draw the density curve 
kde_lines = ax.get_lines()[-1]
kde_x, kde_y = kde_lines.get_data()

# Use Numpy mask to filter the lines for region 
# reresenting height greater than 60 inches 
mask = kde_x > 62
filled_x, filled_y = kde_x[mask], kde_y[mask]

# Shade the partial region 
ax.fill_between(filled_x, y1=filled_y)

# vertical line at x = 60 for reference
plt.axvline(x=62, linewidth=3, linestyle='--')

plt.title("FIG 2. Partial Area: Taller Than 5 Feet", pad=20)
plt.xlabel("Height (inches)", labelpad=20)
plt.ylabel("Probability", labelpad=20)
plt.show()
```

```python
# Area of the orange region (66 inches & above)
area = np.trapz(filled_y, filled_x)

print(f"Probability of heights higher than 62 inches: {area.round(4)}")
```

The total area under the curve equals 1, and the shaded region has an area of 0.9216. Thus it represents 92,16% of all possible height measurements.

Thus, we can say that 92,16% of high school students are taller than 62 height inches.

```python
plt.figure(figsize=(16, 10))

# Plot density curve 
sns.kdeplot(heights['mean'], linewidth=2, fill=True)
ax = sns.kdeplot(heights['mean'])

# Get all the lines used to draw density curve 
kde_lines = ax.get_lines()[-1]
kde_x, kde_y = kde_lines.get_data()

# Filter for height between 5 feet (60 inches) & 6 feet (72 inches)
mask = (kde_x > 62) & (kde_x < 67) 
filled_x, filled_y = kde_x[mask], kde_y[mask]

# Shade the partial region 
ax.fill_between(filled_x, y1=filled_y)

# Vertical lines at 5 and 6 feet for reference
plt.axvline(x=62, linewidth=3, linestyle='--')
plt.axvline(x=67, linewidth=3, linestyle='--')

plt.title("FIG 3. Partial Area: Height Between 5 and 6 Feet", pad=20)
plt.xlabel("Height (inches)", labelpad=20)
plt.ylabel("Probability", labelpad=20)
plt.show()
```

```python
area = np.trapz(filled_y, filled_x)
print(f"Fraction of heights between 62 and 67 feet: {area.round(4)}")
```

# Thus we can conclude that a little over 82% of the high school students are taller than 62 feet but shorter than 67 feet.

```python
plt.figure(figsize=(16, 10))

# Plot density curve 
sns.kdeplot(heights['mean'], linewidth=1, fill=True)
ax = sns.kdeplot(heights['mean'], linewidth=2)

# Get lines used to draw density curve 
kde_lines = ax.get_lines()[-1]
kde_x, kde_y = kde_lines.get_data()

monica_height = 67

# Filter and shade the region with for heights
#  less than monica's 
mask = kde_x < monica_height
filled_x, filled_y = kde_x[mask], kde_y[mask]
ax.fill_between(filled_x, y1=filled_y)

# vertical line for reference
plt.axvline(x=monica_height, linewidth=3, linestyle='--')

plt.title("FIG 4. Partial Area: Shorter Than 67 Inches", pad=20)
plt.xlabel("Height (inches)", labelpad=20)
plt.ylabel("Probability", labelpad=20)
plt.show()
```

```python
# area of the orange region 
area = np.trapz(filled_y, filled_x)
# convert to percentage
pct = (area*100).round(2)
print(f"Percent of heights below {monica_height} inches: {pct}%")
```

# Percentiles concept:

```python
import pandas as pd
import random

A = [ random.randint(0,100) for i in range(10) ]
B = [ random.randint(0,100) for i in range(10) ]

df = pd.DataFrame({ 'field_A': A, 'field_B': B })
df
```

```python
mean = df.field_A.mean()

median = df.field_A.median()

# You can call `quantile(i)` to get the i'th quantile,
# where `i` should be a fractional number.

quantile_10 = df.field_A.quantile(0.1) # 10th percentile


quantile_50 = df.field_A.quantile(0.5) # same as median


quantile_90 = df.field_A.quantile(0.9)# 90th percentile
```

```python
print(f"field_10 value")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"10% percentile: {quantile_10}")
print(f"50% percentile: {quantile_50}")
print(f"90% percentile: {quantile_90}")
```
